```python
def extract_property_details(url):
    df = pd.DataFrame(columns=["Title", "Price", "Property Type", "Bedrooms", "Bathrooms", "Area", "Location"])
    for i in range(290):
        response = requests.get(f"https://www.bayut.sa/en/for-sale/properties/riyadh/page-{i}/")
        soup = BeautifulSoup(response.text, "html.parser")
        properties = soup.find_all("div", class_="_475e888a")
        for prop in properties:
            Title = prop.find("h2", {"aria-label": "Title"}).text.strip() if prop.find("h2", {"aria-label": "Title"}) else "N/A"
            Price = prop.find("span", {"aria-label": "Price"}).text.strip() if prop.find("span", {"aria-label": "Price"}) else "N/A"
            Property_Type = prop.find("span", {"aria-label": "Type"}).text.strip() if prop.find("span", {"aria-label": "Type"}) else "N/A"
            Bedrooms = prop.find("span", {"aria-label": "Beds"}).text.strip() if prop.find("span", {"aria-label": "Beds"}) else "N/A"
            Bathrooms = prop.find("span", {"aria-label": "Baths"}).text.strip() if prop.find("span", {"aria-label": "Baths"}) else "N/A"
            Area = prop.find("h4", class_="cfac7e1b _85ddb82f").text.strip() if prop.find("h4", class_="cfac7e1b _85ddb82f") else "N/A"
            Location = prop.find("h3", class_="_4402bd70").text.strip() if prop.find("h3", class_="_4402bd70") else "N/A"
            
            df.loc[len(df)] = [Title, Price, Property_Type, Bedrooms, Bathrooms, Area, Location]
        print(f"Page {i} scraped successfully.")
        time.sleep(random.uniform(1, 3))# Random sleep to avoid being blocked
    df.to_csv("bayut_properties.csv", index=False)
```
